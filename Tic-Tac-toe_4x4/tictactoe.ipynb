```python
'''
Class 'ExperimentGenerator' which generate a new board game 4x4 has functions:
        1. __init__ : This Function is always required for a class which is used
                      to initialise variables. Here, initBoardState is a variable
                      which is initialized.
        2. generateNewProblem: This Function is used to return a new board state
'''
class ExperimentGenerator():
    ''' Experiment Generator generates New problems. 
    In this case it just returns the same initial game boards '''

    def __init__(self):
        self.initBoardState = [[' ',' ',' ',' '],[' ',' ',' ',' '],[' ',' ',' ',' '],[' ',' ',' ',' ']]

    def generateNewProblem(self):
        ''' Returns initial Board State '''
        return(self.initBoardState)
```

```python
'''
Class 'Player' which provide following required functions to a player:
        1. __init__ : This Function is always required for a class which is used
                      to initialise variables. Here, the variables are playerSymbol 
                      which is initialised with values(i.e symbol of player 'X' or 'O') 
                      while creating object of Player class and playerTargetFunctionWeightVector
                      which is initialised with weight vector of target function.
        2. isGameOver: This Function is used to tell that if the game is over or not
                       if Game is over ,it returns True else return False
        3. lookForLegalMoves: This Function is used to find all legal moves from the
                              current board state.It returns an vector of all board state
                              with next move.
        4. extractFeatures: This Function is use to extract features from the current board
                            state. It returns a features vector after extract all the features.
        5. boardPrint: This Function is use to print the current board state.
        6. calculateNonFinalBoardScore: This Function is use to calculate the score of non 
                                        final board state with the help of weight vector
                                        and feature vector.Then it returns the score.
        7. chooseMove: This Function give player a chance to choose optimal move 
                        from the set of all legal moves. Optimal move is known by calculating
                        the score.It return the optimal move.
        8. chooseRandomMove: This Function use to return the next move for a player
                             which is selected randomly from set of all legal move.
'''
class Player:

    def __init__(self,playerSymbol,playerTargetFunctionWeightVector):
    
        self.playerSymbol = playerSymbol
        self.playerTargetFunctionWeightVector = playerTargetFunctionWeightVector

    def isGameOver(self,board,playerSymbol):
        ''' Returns True if game is over else returns false '''

        flag = False
        # Game already over
        if(board == -1):
            flag = True
        # Game won by either player
        elif((board[0][0] == board[0][1] == board[0][2] == board[0][3] == playerSymbol)  or 
             (board[1][0] == board[1][1] == board[1][2] == board[1][3] == playerSymbol) or
             (board[2][0] == board[2][1] == board[2][2] == board[2][3] == playerSymbol) or
             (board[3][0] == board[3][1] == board[3][2] == board[3][3] == playerSymbol) or
             (board[0][0] == board[1][0] == board[2][0] == board[3][0] == playerSymbol) or
             (board[0][1] == board[1][1] == board[2][1] == board[3][1] == playerSymbol) or
             (board[0][2] == board[1][2] == board[2][2] == board[3][2] == playerSymbol) or
             (board[0][3] == board[1][3] == board[2][3] == board[3][3] == playerSymbol) or
             (board[0][0] == board[1][1] == board[2][2] == board[3][3] == playerSymbol) or
             (board[0][3] == board[1][2] == board[2][1] == board[3][0] == playerSymbol) ): 
                flag = True
        # Board Full
        elif(' ' not in np.array(board).flatten()):
            flag = True
        return(flag)

    def lookForLegalMoves(self,boardState,playerSymbol):
        ''' Returns a list of legal moves for a given board state.'''

        legalMoves = []
        for i in range(len(boardState[0])):
            for j in range(len(boardState[0])):
                if(boardState[i][j] == ' '):
                    tempBoard = copy.deepcopy(boardState)
                    tempBoard[i][j]=playerSymbol
                    legalMoves.append(tempBoard)
        return(legalMoves)

    def extractFeatures(self,board,playerSymbol1,playerSymbol2):
        ''' Returns and extracted feature Vector for a given board state '''

        w1,w2,w3,w4,w5,w6,w7,w8 = 0,0,0,0,0,0,0,0
        
        for i in range(len(board[0])):
            # 1 Player-1's Symbol in a row in an open row
            if (((board[i][0] == playerSymbol1) and ( board[i][1] == board[i][2] == board[i][3] == ' ')) or
                ((board[i][3] == playerSymbol1) and ( board[i][0] == board[i][1] == board[i][2] == ' '))):
                w1 = w1 + 1
            # 1 Player-1's Symbol in a row in an open col
            if (((board[0][i] == playerSymbol1) and ( board[1][i] == board[2][i] == board[3][i] == ' ')) or
                ((board[3][i] == playerSymbol1) and ( board[0][i] == board[1][i] == board[2][i] == ' '))):
                w1 = w1 + 1
            # 1 Player-2's Symbol in a row in an open row
            if (((board[i][0] == playerSymbol2) and ( board[i][1] == board[i][2] == board[i][3] == ' ')) or
                ((board[i][3] == playerSymbol2) and ( board[i][0] == board[i][1] == board[i][2] == ' '))):
                w2 = w2 + 1
            # 1 Player-2's Symbol in a row in an open col
            if (((board[0][i] == playerSymbol2) and ( board[1][i] == board[2][i] == board[3][i] == ' ')) or
                ((board[3][i] == playerSymbol2) and ( board[0][i] == board[1][i] == board[2][i] == ' '))):
                w2 = w2 + 1
            # 2 Player-1's Symbols in a row
            if (((board[i][0] == board[i][1] == playerSymbol1) and (board[i][2] == board[i][3] == ' ')) or
                ((board[i][2] == board[i][3] == playerSymbol1) and (board[i][0] == board[i][1] == ' '))):
                w3 = w3 + 1
            # 2 Player-2's Symbols in a row
            if (((board[i][0] == board[i][1] == playerSymbol2) and (board[i][2] == board[i][3] == ' ')) or
                ((board[i][2] == board[i][3] == playerSymbol2) and (board[i][0] == board[i][1] == ' '))):
                w4 = w4 + 1
            # 3 Player-1's Symbols in a row
            if (((board[i][0] == board[i][1] == board[i][2] == playerSymbol1) and (board[i][3] == ' ')) or
                ((board[i][1] == board[i][2] == board[i][3] == playerSymbol1) and (board[i][0] == ' '))):
                w5 = w5 + 1
            # 3 Player-2's Symbols in a row
            if (((board[i][0] == board[i][1] == board[i][2] == playerSymbol2) and (board[i][3] == ' ')) or
                ((board[i][1] == board[i][2] == board[i][3] == playerSymbol2) and (board[i][0] == ' '))):
                w6 = w6 + 1
            # 4 Player-1's Symbols in a row with an open box
            if ((board[i][0] == board[i][1] == board[i][2] == board[i][3] == playerSymbol1)):
                w7 = w7 + 1
            # 4 Player-2's Symbols in a row with an open box
            if (board[i][0] == board[i][1] == board[i][2] == board[i][3] == playerSymbol2 ):
                w8 = w8 + 1
        # Added 1 for bias
        feature_vector = [1,w1,w2,w3,w4,w5,w6,w7,w8]
        return(feature_vector)
        
    def boardPrint(self,board):
        ''' Displays board in proper format'''
        print('\n')
        print("---------")
        for i in range(len(board[0])):
            st = "|"
            for j in range(len(board[0])):
                st = st + board[i][j] + "|"
            print(st)
            print("---------")
        print('\n')

    def calculateNonFinalBoardScore(self,weight_vector,feature_vector):
        ''' Returns score/Value of a given non final board state '''

        weight_vector = np.array(weight_vector).reshape((len(weight_vector),1))
        feature_vector = np.array(feature_vector).reshape((len(feature_vector),1))
        boardScore = np.dot(weight_vector.T,feature_vector)
        return(boardScore[0][0])
    
    def chooseMove(self,board,playerSymbol1,playerSymbol2):
        ''' Returns the best move from a set of legal moves for a given board state.'''

        legalMoves = self.lookForLegalMoves(board,playerSymbol1)
        legalMoveScores = [self.calculateNonFinalBoardScore(self.playerTargetFunctionWeightVector,
            self.extractFeatures(i,playerSymbol1,playerSymbol2)) for i in legalMoves]
        newBoard = legalMoves[np.argmax(legalMoveScores)]
        return(newBoard)

    def chooseRandomMove(self,board,playerSymbol):
        ''' Returns a random move from a set of legal moves for a given board state '''
        legalMoves = self.lookForLegalMoves(board,playerSymbol)
        newBoard = random.choice(legalMoves)
        return(newBoard)
```

```python
''' 
class 'Performance System' used to make Gamehistory which has following functions:
        1. __init__ : This Function is always required for a class which is used
                      to initialise variables.Here, it initialises the variables
                      board with initial board state,playersTargetFunctionWeightVectors
                      with weight vector of target function,and playerSymbols- an array
                      which contain symbols for both players.
        2. isGameOver: This Function is used to tell that if the game is over or not
                       if Game is over ,it returns True else return False.
        3. generateGameHistory: This Function is use to create game history which consist
                                of all the moves done by players till any of them wins or
                                its draw.In this function two objects of players are created.
                                After that loop is run till the game is over.Then it returns 
                                the game History.
'''
class PerformanceSystem:

    def __init__(self,initialBoard,playersTargetFunctionWeightVectors,playerSymbols):

        self.board = initialBoard
        self.playersTargetFunctionWeightVectors = playersTargetFunctionWeightVectors
        self.playerSymbols = playerSymbols

    def isGameOver(self,board,playerSymbol):
        ''' Returns True if game is over else returns false '''

        flag = False
        # Game already over
        if(board == -1):
            flag = True
        # Game won by either player
        elif((board[0][0] == board[0][1] == board[0][2] == board[0][3] == playerSymbol)  or 
             (board[1][0] == board[1][1] == board[1][2] == board[1][3] == playerSymbol) or
             (board[2][0] == board[2][1] == board[2][2] == board[2][3] == playerSymbol) or
             (board[3][0] == board[3][1] == board[3][2] == board[3][3] == playerSymbol) or
             (board[0][0] == board[1][0] == board[2][0] == board[3][0] == playerSymbol) or
             (board[0][1] == board[1][1] == board[2][1] == board[3][1] == playerSymbol) or
             (board[0][2] == board[1][2] == board[2][2] == board[3][2] == playerSymbol) or
             (board[0][3] == board[1][3] == board[2][3] == board[3][3] == playerSymbol) or
             (board[0][0] == board[1][1] == board[2][2] == board[3][3] == playerSymbol) or
             (board[0][3] == board[1][2] == board[2][1] == board[3][0] == playerSymbol) ): 
                flag = True
        # Board full
        elif(' ' not in np.array(board).flatten()):
            flag = True
        return(flag)

    def generateGameHistory(self):
        ''' Returns Solution trace generated from pitting 2 players(agents) against each '''

        gameHistory = []
        gameStatusFlag = True
        player1 = Player(self.playerSymbols[0],self.playersTargetFunctionWeightVectors[0])
        player2 = Player(self.playerSymbols[1],self.playersTargetFunctionWeightVectors[1])
        tempBoard = copy.deepcopy(self.board)
        while(gameStatusFlag):
            #tempBoard = player1.chooseRandomMove(tempBoard,player1.playerSymbol)
            tempBoard = player1.chooseMove(tempBoard,player1.playerSymbol,player2.playerSymbol)
            gameHistory.append(tempBoard)
            gameStatusFlag = not self.isGameOver(tempBoard,player1.playerSymbol)
            if(gameStatusFlag == False):
                break
            tempBoard = player2.chooseRandomMove(tempBoard,player2.playerSymbol)
            #tempBoard = player2.chooseMove(tempBoard,player2.playerSymbol,player1.playerSymbol)
            gameHistory.append(tempBoard)
            gameStatusFlag =  not self.isGameOver(tempBoard,player2.playerSymbol)
        return(gameHistory)
```

```python
'''
class 'Critic' uses the game history to make training examples with help of following functions:
        1. __init__ : This Function is always required for a class which is used
                      to initialise variables.Here, variable gameHistory is initialized with
                      the history of moves that PerformanceSystem has made.
        2. extractFeatures: This Function is use to extract features from the current board
                            state. It returns a features vector after extracting all the features.
        3. calculateNonFinalBoardScore: This Function is use to calculate the score of non 
                                        final board state with the help of weight vector
                                        and feature vector.Then it returns the score.
        4. calculateFinalBoardScore: This Function is use to calculate the final board score by
                                     giving 100 score if player 1 wins ,if it loses then -100
                                     else 0 for draw.Then, it returns the score.
        5. generateTrainingSamples: This Function generate training sample for training furthur
                                    by storing feature vector with its non final score and feature
                                    vector of non-final board state and its score.Then, it returns 
                                    the training sample vector.
        6. arrayPrint: This function is used to print board state.
        7. boardDisplay: This function is used to print which player won , and also it store 
                         the count of how many p1 won,p2 won and draw matches.Then its return
                         that count which is stored in gameStatusCount.
            
'''
class Critic:
    ''' Critic takes the Game History & generates training examples to be used by Generalizer.'''

    def __init__(self,gameHistory):
        self.gameHistory = gameHistory

    def extractFeatures(self,board,playerSymbol1,playerSymbol2):
        ''' Returns and extracted feature Vector for a given board state '''

        w1,w2,w3,w4,w5,w6,w7,w8 = 0,0,0,0,0,0,0,0
        
        for i in range(len(board[0])):
            # 1 Player-1's Symbol in a row in an open row
            if (((board[i][0] == playerSymbol1) and ( board[i][1] == board[i][2] == board[i][3] == ' ')) or
                ((board[i][3] == playerSymbol1) and ( board[i][0] == board[i][1] == board[i][2] == ' '))):
                w1 = w1 + 1
            # 1 Player-1's Symbol in a row in an open col
            if (((board[0][i] == playerSymbol1) and ( board[1][i] == board[2][i] == board[3][i] == ' ')) or
                ((board[3][i] == playerSymbol1) and ( board[0][i] == board[1][i] == board[2][i] == ' '))):
                w1 = w1 + 1
            # 1 Player-2's Symbol in a row in an open row
            if (((board[i][0] == playerSymbol2) and ( board[i][1] == board[i][2] == board[i][3] == ' ')) or
                ((board[i][3] == playerSymbol2) and ( board[i][0] == board[i][1] == board[i][2] == ' '))):
                w2 = w2 + 1
            # 1 Player-2's Symbol in a row in an open col
            if (((board[0][i] == playerSymbol2) and ( board[1][i] == board[2][i] == board[3][i] == ' ')) or
                ((board[3][i] == playerSymbol2) and ( board[0][i] == board[1][i] == board[2][i] == ' '))):
                w2 = w2 + 1
            # 2 Player-1's Symbols in a row
            if (((board[i][0] == board[i][1] == playerSymbol1) and (board[i][2] == board[i][3] == ' ')) or
                ((board[i][2] == board[i][3] == playerSymbol1) and (board[i][0] == board[i][1] == ' '))):
                w3 = w3 + 1
            # 2 Player-2's Symbols in a row
            if (((board[i][0] == board[i][1] == playerSymbol2) and (board[i][2] == board[i][3] == ' ')) or
                ((board[i][2] == board[i][3] == playerSymbol2) and (board[i][0] == board[i][1] == ' '))):
                w4 = w4 + 1
            # 3 Player-1's Symbols in a row
            if (((board[i][0] == board[i][1] == board[i][2] == playerSymbol1) and (board[i][3] == ' ')) or
                ((board[i][1] == board[i][2] == board[i][3] == playerSymbol1) and (board[i][0] == ' '))):
                w5 = w5 + 1
            # 3 Player-2's Symbols in a row
            if (((board[i][0] == board[i][1] == board[i][2] == playerSymbol2) and (board[i][3] == ' ')) or
                ((board[i][1] == board[i][2] == board[i][3] == playerSymbol2) and (board[i][0] == ' '))):
                w6 = w6 + 1
            # 4 Player-1's Symbols in a row with an open box
            if ((board[i][0] == board[i][1] == board[i][2] == board[i][3] == playerSymbol1)):
                w7 = w7 + 1
            # 4 Player-2's Symbols in a row with an open box
            if (board[i][0] == board[i][1] == board[i][2] == board[i][3] == playerSymbol2 ):
                w8 = w8 + 1
        # Added 1 for bias
        feature_vector = [1,w1,w2,w3,w4,w5,w6,w7,w8]
        return(feature_vector)



    def calculateNonFinalBoardScore(self,weight_vector,feature_vector):
        ''' Returns score/Value of a given non final board state '''

        weight_vector = np.array(weight_vector).reshape((len(weight_vector),1))
        feature_vector = np.array(feature_vector).reshape((len(feature_vector),1))
        boardScore = np.dot(weight_vector.T,feature_vector)
        return(boardScore[0][0])

    def calculateFinalBoardScore(self,board,playerSymbol1,playerSymbol2):
        ''' Returns score/Value of a given final board state '''

        # If game ends in a draw
        score = 0
        # If player-1 (i.e self) wins
        if((board[0][0] == board[0][1] == board[0][2] == board[0][3] == playerSymbol1)  or 
            (board[1][0] == board[1][1] == board[1][2] == board[1][3] == playerSymbol1) or
            (board[2][0] == board[2][1] == board[2][2] == board[2][3] == playerSymbol1) or
            (board[3][0] == board[3][1] == board[3][2] == board[3][3] == playerSymbol1) or
            (board[0][0] == board[1][0] == board[2][0] == board[3][0] == playerSymbol1) or
            (board[0][1] == board[1][1] == board[2][1] == board[3][1] == playerSymbol1) or
            (board[0][2] == board[1][2] == board[2][2] == board[3][2] == playerSymbol1) or
            (board[0][3] == board[1][3] == board[2][3] == board[3][3] == playerSymbol1) or
            (board[0][0] == board[1][1] == board[2][2] == board[3][3] == playerSymbol1) or
            (board[0][3] == board[1][2] == board[2][1] == board[3][0] == playerSymbol1) ):
            score = 100
        # If player-2 (i.e opponent) wins
        elif((board[0][0] == board[0][1] == board[0][2] == board[0][3] == playerSymbol2)  or 
            (board[1][0] == board[1][1] == board[1][2] == board[1][3] == playerSymbol2) or
            (board[2][0] == board[2][1] == board[2][2] == board[2][3] == playerSymbol2) or
            (board[3][0] == board[3][1] == board[3][2] == board[3][3] == playerSymbol2) or
            (board[0][0] == board[1][0] == board[2][0] == board[3][0] == playerSymbol2) or
            (board[0][1] == board[1][1] == board[2][1] == board[3][1] == playerSymbol2) or
            (board[0][2] == board[1][2] == board[2][2] == board[3][2] == playerSymbol2) or
            (board[0][3] == board[1][3] == board[2][3] == board[3][3] == playerSymbol2) or
            (board[0][0] == board[1][1] == board[2][2] == board[3][3] == playerSymbol2) or
            (board[0][3] == board[1][2] == board[2][1] == board[3][0] == playerSymbol2) ):
            score = -100
        return(score)


    def generateTrainingSamples(self,weight_vector,playerSymbol1,playerSymbol2):
        ''' Returns training examples i.e a list of list of feature vectors & assosiated scores '''

        trainingExamples=[]
        for i in range(len(self.gameHistory)-1):
            feature_vector = self.extractFeatures(self.gameHistory[i+1],playerSymbol1,playerSymbol2)
            trainingExamples.append([feature_vector,self.calculateNonFinalBoardScore(weight_vector,feature_vector)])
        trainingExamples.append([self.extractFeatures(self.gameHistory[-1],playerSymbol1,playerSymbol2),
            self.calculateFinalBoardScore(self.gameHistory[-1],playerSymbol1,playerSymbol2)])
        return(trainingExamples)

    def arrayPrint(self,board):
        ''' Displays a given 2D array in proper format '''
        print('\n')
        print("---------")
        for i in range(len(board[0])):
            st = "|"
            for j in range(len(board[0])):
                st = st + board[i][j] + "|"
            print(st)
            print("---------")
        print('\n')

    def boardDisplay(self,playerSymbol1,playerSymbol2,gameStatusCount):
        ''' Displayes board & returns a list containg Win/Loss/Draw counts '''

        for board in self.gameHistory:
            self.arrayPrint(board)
        finalScore = self.calculateFinalBoardScore(self.gameHistory[-1],playerSymbol1,playerSymbol2)
        if(finalScore == 100):
            print(playerSymbol1 + " wins")
            gameStatusCount[0] = gameStatusCount[0] + 1
        elif(finalScore == -100):
            print(playerSymbol2 + " wins")
            gameStatusCount[1] = gameStatusCount[1] + 1
        else:
            print("Draw")
            gameStatusCount[2] = gameStatusCount[2] + 1
        return(gameStatusCount)
```

```python
'''
class 'Generalizer' takes trining example generated from critic and uses
least means square update rule to learn weights.It has following methods:
        1. __init__ : This Function is always required for a class which is used
                      to initialise variables.Here, variable trainingExamples is 
                      initialised with training examples prepared by critic.
        2. calculateNonFinalBoardScore: This Function is use to calculate the score of non 
                                        final board state with the help of weight vector
                                        and feature vector.Then it returns the score.
        3. lmsWeightUpdate: This Function uses least mean square update rule to
                            update weights
'''
class Generalizer:

    def __init__(self,trainingExamples):
        self.trainingExamples = trainingExamples

    def calculateNonFinalBoardScore(self,weight_vector,feature_vector):
        ''' Returns score/Value of a given non final board state '''

        weight_vector = np.array(weight_vector).reshape((len(weight_vector),1))
        feature_vector = np.array(feature_vector).reshape((len(feature_vector),1))
        boardScore = np.dot(weight_vector.T,feature_vector)
        return(boardScore[0][0])

    def lmsWeightUpdate(self,weight_vector,alpha=0.4):
        ''' Returns new Weight vector updated y learning from Training 
            examples via LMS (Least Mean Squares) training rule.'''

        for trainingExample in self.trainingExamples:
            vTrainBoardState = trainingExample[1]
            vHatBoardState = self.calculateNonFinalBoardScore(weight_vector,trainingExample[0])
            weight_vector = weight_vector + (alpha * (vTrainBoardState - vHatBoardState) * np.array(trainingExample[0]))
        return (weight_vector)
```

<center><h1>********** Training Phase **********</h1></center>

```python
'''
Function: train_TTT_Computer
Inputs: number of training sample which is initialised to one
Reason: This Function is used to learn the Weight vector for the tic-tac-toe
        computer game, with the help of experiment generator which generate 
        new boad state,then performance system which gives game history for 
        each game,then critic which is used to generate training example for 
        each player and finally generalizer is used to update weights using 
        LMS rule of update.And Finally it prints the learnt weights.
'''
def train_TTT_Computer(numTrainingSamples = 10):
    ''' Executions of Training & Testing phases'''

    # Training phase (Indirect Feedback via Computer v/s Computer)
    trainingGameCount = 0
    playerSymbols = ('X','O')
    playersTargetFunctionWeightVectors = [np.array([.5,.5,.5,.5,.5,.5,.5,.5,.5]),np.array([.5,.5,.5,.5,.5,.5,.5,.5,.5])]
    gameStatusCount = [0,0,0]

    while (trainingGameCount < numTrainingSamples):

        # Experiment Generator
        experimentGenerator = ExperimentGenerator()
        initialBoardState = experimentGenerator.generateNewProblem()

        # Performance System
        performanceSystem = PerformanceSystem(initialBoardState,playersTargetFunctionWeightVectors,playerSymbols)
        gameHistory = performanceSystem.generateGameHistory()

        # Critic
        critic = Critic(gameHistory)
        trainingExamplesPlayer1 = critic.generateTrainingSamples(playersTargetFunctionWeightVectors[0],
            playerSymbols[0],playerSymbols[1])
        trainingExamplesPlayer2 = critic.generateTrainingSamples(playersTargetFunctionWeightVectors[1],
            playerSymbols[1],playerSymbols[0])
        # Display board states
        gameStatusCount = critic.boardDisplay(playerSymbols[0],playerSymbols[1],gameStatusCount)

        # Generalizer
        generalizer = Generalizer(trainingExamplesPlayer1)
        playersTargetFunctionWeightVectors = [generalizer.lmsWeightUpdate(playersTargetFunctionWeightVectors[0]),
            generalizer.lmsWeightUpdate(playersTargetFunctionWeightVectors[1])]

        trainingGameCount = trainingGameCount+1

    # Epoch wise Game status
    print("\nTraining Results: (" + "Player-1 Wins = " + str(gameStatusCount[0]) +
            ", Player-2 Wins = " + str(gameStatusCount[1]) + ", Game Draws = " + str(gameStatusCount[2]) +
            ")\n")	

    # Weight Learnt from previous games
    learntWeight =  list(np.mean(np.array([playersTargetFunctionWeightVectors[0],
        playersTargetFunctionWeightVectors[1]]),axis = 0))
    return learntWeight
```

Here, we are training for small sample only because while making pdf ,it will require large number of pages, so training with less number of samples.

```python
'''
Function: playAgainstComputer
Inputs: FinalWeight
Reason: This Function is used for testing manualy by playing against 
        learnt TTT_computer.
'''
def playAgainstComputer(FinalWeight):
    # Computer vs Human Games
    ans = "y"
    while(ans == "y"):

        experimentGenerator = ExperimentGenerator()
        boardState = experimentGenerator.generateNewProblem()
        gameStatusFlag = True
        computer = Player('X',FinalWeight)
        gameHistory = []

        print('\nBegin Computer(X) v/s Human(O) Tic-Tac-Toe\n')
        while(gameStatusFlag):

            boardState = computer.chooseMove(boardState,computer.playerSymbol,'O')
            print('Computers\'s Turn:\n')
            computer.boardPrint(boardState)
            gameHistory.append(boardState)
            gameStatusFlag = not computer.isGameOver(boardState,computer.playerSymbol)
            if(gameStatusFlag == False):
                print("\n\n OOPS!.... You lost by this computer :( \n\n")
                break

            print('Human\'s Turn:\n')
            while(True):
                print('Enter X-coordinate(0-3):')
                x = int(input())
                print('Enter Y-coordinate(0-3):')
                y = int(input())
                if((x > 3) or (x < 0) or (y < 0) or (y > 3) or boardState[x][y] != ' ' ):
                    print("Please Enter again (coordinate u entered may out of box or already filled) ")
                else:
                    break

            boardState[x][y] = 'O'
            gameStatusFlag = not computer.isGameOver(boardState,'O')
            if gameStatusFlag == False:
                print("\n\n****Congratulations!!!, You defeated Computer :) ****\n\n ")
                break
            computer.boardPrint(boardState)
            gameHistory.append(boardState)

        print("Do you want to continue playing(y/n).")
        ans = input()
        if(ans != 'y'):
            break
```

```python
# Calling function to play against learnt computer. 
playAgainstComputer(FinalWeight)
```
